```python
df['INSTRUCTION_EXPENDITURE_PERCENT_TOTAL'] = df['INSTRUCTION_EXPENDITURE'] / df['TOTAL_EXPENDITURE'] * 100
```

Feature engineering justification: Wanted to see how the ratio of instruction expenditures to total expenditures affected test scores.

<h2>Visualization</h2>

Investigate the relationship between your chosen response variable and at least two predictors using visualizations. Write down your observations.

**Visualization 1**

```python
df.plot.scatter(x='INSTRUCTION_EXPENDITURE_PERCENT_TOTAL', y='AVG_MATH_8_SCORE', alpha=0.6)
plt.xlabel('Instruction Expenditure Percent of Total Expenditure')
plt.ylabel('8th Grade Math Score')
```

The percentage of total expenditure that instruction expenditure takes up, overall, does not generally affect the average 8th-grade math score.

**Visualization 2**

```python
df.plot.scatter(x='AVG_MATH_4_SCORE', y='AVG_MATH_8_SCORE', alpha=0.6)
plt.xlabel('4th Grade Math Score')
plt.ylabel('8th Grade Math Score')
```

This shows that 4th-grade scores tend to correlate with 8th-grade scores and that relatively speaking, kids do not improve  between these ages.

<h2> Data Creation </h2>

_Use this space to create train/test data_

```python
from sklearn.model_selection import train_test_split
```

```python
X = df[['INSTRUCTION_EXPENDITURE_PERCENT_TOTAL','AVG_MATH_4_SCORE','AVG_READING_8_SCORE']].dropna()
y = df.loc[X.index]['AVG_MATH_8_SCORE'] 
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.4, random_state=42)
```

<h2> Prediction </h2>

ML Models [Resource](https://medium.com/@vijaya.beeravalli/comparison-of-machine-learning-classification-models-for-credit-card-default-data-c3cf805c9a5a)

```python
from sklearn.linear_model import LinearRegression
```

```python
# create your model here 
model = LinearRegression()
```

```python
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

We have copied over the graphs that visualize the model's performance on the training and testing set. 

Change `col_name` and modify the call to `plt.ylabel()` to isolate how a single predictor affects the model.

```python
col_name = 'INSTRUCTION_EXPENDITURE_PERCENT_TOTAL'

f = plt.figure(figsize=(12,6))
plt.scatter(X_train[col_name], y_train, color = "red")
plt.scatter(X_train[col_name], model.predict(X_train), color = "green")

plt.legend(['True Training','Predicted Training'])
plt.xlabel(col_name)
plt.ylabel('Average 8th Grade Math Score')
plt.title("Model Behavior On Training Set")
```

```python
col_name = 'AVG_MATH_4_SCORE'

f = plt.figure(figsize=(12,6))
plt.scatter(X_test[col_name], y_test, color = "blue")
plt.scatter(X_test[col_name], model.predict(X_test), color = "black")

plt.legend(['True testing','Predicted testing'])
plt.xlabel(col_name)
plt.ylabel('Average 8th Grade Math Score')
plt.title("Model Behavior on Testing Set")
```

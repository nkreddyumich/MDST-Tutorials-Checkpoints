```python
# load necessary libraries
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
import sklearn
import pandas as pd
import keras
from keras.models import Sequential
from keras.layers import Dense
import tensorflow as tf
from sklearn.metrics import confusion_matrix,ConfusionMatrixDisplay
from sklearn.preprocessing import MinMaxScaler
```

```python
# read in data
df_temp = pd.read_csv("/content/drive/MyDrive/startup data.csv")
df = df_temp.copy()
# explore data
df.head()
```

```python
# convert non-numeric values to NaN
df = df.apply(pd.to_numeric, errors = 'coerce')
# drop rows with NaN values
df = df.dropna(axis = 1)
```

```python
# check how many input columns
df.head()
```

```python
# split input and output
target_variable = 'labels'
new_df = df.columns.drop(target_variable)
rows = df[target_variable]
df.drop(target_variable, axis = 1, inplace = True)
```

```python
# apply min-max normalization
scaler = MinMaxScaler()
scaler.fit(df)
transformed_df = scaler.transform(df)
```

```python
# split dataset into training and testing sets (70% train, 30% test)
X_train, X_test, y_train, y_test = train_test_split(transformed_df, rows, test_size = 0.30, random_state = 0)
```

```python
# build ML model
model = Sequential()
model.add(Dense(units = 16, activation = 'relu', input_shape = (32,)))
model.add(Dense(1, activation = 'sigmoid'))
```

```python
# initialize optimizer for training
opt = keras.optimizers.Adam(learning_rate = 0.001)
# compile model
model.compile(loss = 'binary_crossentropy', optimizer = opt, metrics=["accuracy"])
```

```python
# train model
model.fit(X_train, y_train, epochs = 100)
```

```python
# test and evaluate accuracy of the model
test_results = model.evaluate(X_test, y_test)
print('Accuracy = ', test_results[1] * 100, '%')
```
